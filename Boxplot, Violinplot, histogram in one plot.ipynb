```python
df=pd.read_csv('DATA/Telco-Customer-Churn.csv')
df=df.drop('customerID',axis=1)
cat_col=df.select_dtypes(include='object')
num_col=df.select_dtypes(include=['float64','int64'])
```

```python
for i in num_col:
    
    fig,ax=plt.subplots(nrows=1,ncols=3,
                        sharey=True,
                        figsize=(12,4))
  
    sns.boxplot(data=df,x='Churn',y=i,ax=ax[0])
    sns.violinplot(data=df,x='Churn',y=i,ax=ax[1])
    sns.histplot(data=df,y=i,ax=ax[2],bins=30)
    
    plt.tight_layout()
    plt.show()
```
